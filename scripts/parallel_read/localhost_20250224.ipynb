```python
import pandas as pd

threads = [2, 4, 6, 8, 10]
latencies = [245, 153, 135, 119, 115]
data_size = 8450

df = pd.DataFrame({'thread': threads, 'latency': latencies})
df['throughput'] = data_size / df['latency']

df.plot(x='thread', y='throughput', xlabel='Количество читающих потоков kqprun', ylabel='Пропускная способность, МБ/с', legend=False, 
        title='Скорость чтения через коннектор\nколоночной таблицы YDB TPC-H S10 Lineitem')
```
